```python
# Testing

size=10000
mean=10
sd=10

a = np.random.normal(mean,sd,size)
# a = np.log2(a)
a = (a-mean)/sd

def quantile_calculator(a):
    quantile = []
    for i in range(len(a)):
        q = (i-0.5)/len(a)
        q = norm.ppf(q,loc=0,scale=1)
        quantile.append(q)
    return quantile
quantile = quantile_calculator(a)
a = np.sort(a)
plt.scatter(quantile,a)
# plt.hist(a)
```

```python
# Read data
data = pd.read_csv("Cholesterol.csv",index_col=False)

# Take the natural log of the number
# data = np.log(data)

# Normalize data
mean = data.mean()
sd = data.std()
data = (data-mean)/sd

# Rank the data
data = data.sort_values(by="value")

# Produce quantile
def quantile_calculator(a):
    quantile = []
    for i in range(len(a)):
        q = (i-0.5)/len(a)
        q = norm.ppf(q,loc=0,scale=1)
        quantile.append(q)
    return quantile
quantile = quantile_calculator(data)

# Plot
plt.scatter(quantile,data)
```
