```python
# Read data
data = pd.read_csv("data_change_heart_disease_numerical.csv")
data = data.drop("HeartDisease",axis=1)

# Take the natural log of the number
# data = np.log(data)

# Normalize data
def sample_quantile(df):
    df = df.dropna()
    mean = df.mean()
    sd = df.std()
    df = (df-mean)/sd
    df = df.sort_values()
    return df

def theoretical_quantile(a):
    a = a.dropna()
    quantile = []
    for i in range(len(a)):
        q = (i-0.5)/len(a)
        q = norm.ppf(q,loc=0,scale=1)
        quantile.append(q)
    quantile = pd.Series(quantile)
    return quantile

def draw_graph(df,ax,i):
    df=data.loc[:,i]
    sam_quan = sample_quantile(df)
    the_quan = theoretical_quantile(df)
    # a, b = np.polyfit(the_quan, sam_quan, 1)
    # ax1.plot(the_quan,the_quan*the_quan+b, color='r')
    ax.scatter(the_quan,sam_quan,s=2,c='#007789')
    ax.set_title(i)
    ax.set_xlabel("Theoretical quantile",labelpad=None )
    ax.set_ylabel("Sample quantile")
    # return ax

# for i in data.columns:
#     df = data[i]
#     sam_quan = sample_quantile(df)
#     the_quan = theoretical_quantile(df)
    

# Rank the data


# Produce quantile

quantile = theoretical_quantile(data)

# Plot
# plt.scatter(quantile,data)
```
